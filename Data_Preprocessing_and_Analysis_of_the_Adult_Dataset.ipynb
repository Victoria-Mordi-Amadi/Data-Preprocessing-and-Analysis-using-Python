```python

def rearrange(df):
    #Finding midpoint
    mid = len(df)//2

    #Splitting in the middle
    part_A = df.iloc[:mid]
    part_B = df.iloc[mid:]

    #Resetting the table
    rearranged = pd.concat([part_B, part_A], ignore_index = True)

    return rearranged

# Applying function
rearrange(task_dataset)
```

```python

def compare_col(df, col_A, col_B):
    # Creating a copy of the original data
    df_I = df.copy(deep = True)

    # Compare the values of first and second columns for each row
    comparison = df_I[col_A] > df_I[col_B]
    # Converting result to list
    comparison = comparison.tolist()
    # Making the list into a dataframe
    comparison_df = pd.DataFrame(comparison, columns = ['Greater'])

    #Resetting index
    df_I = df_I.reset_index(drop=True)
    comparison_df = comparison_df.reset_index(drop=True)

    #Concatenating the tables
    df_II = pd.concat([df_I, comparison_df], ignore_index = True, axis = 1)

    return df_II

New_task = pd.DataFrame(task_dataset[['hours-per-week', 'age']])

# Getting comparisons
Comparison = compare_col(New_task,'hours-per-week', 'age')

# Renaming columns
Comparison.rename(columns = {0: 'Hours/week',
                           1: 'Age',
                           2: 'Greater'}, inplace = True)

# Print the comparison result
print(Comparison)
```

```python
# Calculate average ages for each country

def get_countries_avg_ages(df):

    avg_age_by_country = df.groupby('native-country')['age'].mean()

    # Get the country with maximum average age
    max_avg_age_by_country = avg_age_by_country.idxmax()

    # Get the country with minimum average age
    min_avg_age_by_country = avg_age_by_country.idxmin()

    return max_avg_age_by_country, min_avg_age_by_country

# Get countries with maximum and minimum average ages

max_country, min_country = get_countries_avg_ages(task_dataset)
print("Country with maximum average age:", max_country)
print("Country with minimum average age:", min_country)
```
